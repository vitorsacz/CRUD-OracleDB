### **2 - Criando as tabelas no banco**

 necessrio que voc tenha a tabela criada no seu banco com produtos dentro dela.

Para isso, voc pode usar o arquivo *criar_tabelas.sql* que est na pasta *scripts* do projeto.



__1-__ Copie o Create Table e cole no Sql developer



__2-__ Clique no boto play, no canto superior esquerdo do script



__3-__ Logo aps voc precisar clicar no boto para commitar as alteraes



__4-__ Agora  necessrio inserir os produtos na tabela que criamos, para isso, 
        
        1- copie todos os inserts que esto no arquivo scripts/criar_tabelas.sql , cole no banco de dados e repita os passos 2 e 3 novamente para inserir os dados.

__5-__ Voc pode conferir se os produtos foram adicionados, executando a query *SELECT*, contida na pasta __scripts/criar_tabelas.sql__ e repetir o passo 2.

__6-__ Com isso seu banco de dados est pronto para voce poder realizar o crud.
